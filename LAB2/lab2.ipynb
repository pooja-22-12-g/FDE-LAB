```python

import pandas as pd
df = pd.read_csv(path)
display(df)
```

```python


df1 = df.drop_duplicates()
display(df1)
```

```python
un = df1.fillna("Unknown")

display(un)
```

```python
un['customer_name'] = un['customer_name'].astype(str).str.upper()
display(un)
```

```python
import numpy as np

conditions = [
    df1['annual_spend'] < 80000,
    (df1['annual_spend'] >= 80000) & (df1['annual_spend'] <= 120000)
]
choices = ['low', 'medium']

df1['spend_category'] = np.select(conditions, choices, default='high')
display(df1)


```

```python
customer_summary = df1.groupby(['city', 'spend_category']).agg(
    total_customers=('customer_id', 'size'),
    avg_annual_spend=('annual_spend', 'mean')
).reset_index()
display(customer_summary)
```
